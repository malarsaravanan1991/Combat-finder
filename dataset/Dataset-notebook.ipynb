# COMBAT FINDER - A Fight Detection Dataset




 


### Context


The data set is created for the project " combat finder" which detects fight in the still images and then move the combat finder to next level of video detection with temporal dimension in context.
There was no public data set with labels fight /hit/kick or that has any aggressive action detected in it. 

The fight dataset is created from HMDB51 dataset which is a video dataset. 

So i basically extracted frames from the videos from the classes mentioned below and resized them into images 224 x 224 x 3 where 3 is the RGB channels and 224 is the height and width of the pixel. Also i took 8% of the entire data into testing set and other 92% into training data set.  Here the labels 1,2,3,6 are considered as action related to fight and all other comes under non-fight.

```python
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
import csv
import seaborn as sns
```

### Plotting the total number of examples or data frames in each class/labels used

```python
#adjust the font size in the upcoming plot
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 20}

plt.rcParams.update({'font.size': 20})
matplotlib.rc('font', size=20)
matplotlib.rc('axes', titlesize=20)

#open the csv file to calculate the total images in each classes and plot a bar plot to visualise the
#how the data is spread across different labels
with open("dataset.csv", "r") as f_input:
    csv_input = csv.DictReader(f_input)
    names = []
    num_hit_examples=0
    num_kick_examples =0
    num_punch_examples=0
    num_push_examples=0
    num_ridehorse_examples=0
    num_shootgun_examples=0
    num_stand_examples=0
    num_wave_examples = 0
    for row in csv_input:
        (c) =(row['LabelID'])
        num = int(c)
        if num == 1:
            num_hit_examples +=1
        elif num ==2:
            num_kick_examples+=1
        elif num ==3:
            num_punch_examples+=1
        elif num ==4:
            num_push_examples+=1
        elif num ==5:
            num_ridehorse_examples+=1
        elif num ==6:
            num_shootgun_examples+=1
        elif num ==7: 
            num_stand_examples+=1
        else:
            num_wave_examples+=1
data = [num_hit_examples,num_kick_examples,num_punch_examples,num_push_examples,num_ridehorse_examples,num_shootgun_examples,num_stand_examples,num_wave_examples]
Labels = [1,2,3,4,5,6,7,8]
plt.figure(1)                # the first figure
plt.bar(Labels,data)
plt.ylabel('Number of examples')
plt.xlabel('Classes')
plt.title('# of examples/ class')
plt.show()
```

In above plot the the image frames corresponding to class 1,2,3,6 refer to fight related actions and all other images from 4,5,7,8 refer to non fight. I have added more data so that the model doesn't end up overfitting.
